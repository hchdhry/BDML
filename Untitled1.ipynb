```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, month, count
import matplotlib.pyplot as plt

```

```python
spark = SparkSession.builder.appName("CrimeAnalysis").getOrCreate()
```

```python
crime_data = spark.read.option("header", True).option("inferSchema", True).csv("/content/CrimeData/*.csv")

```

```python
from pyspark.sql.functions import input_file_name, when

crime_data = crime_data.withColumn("file_path", input_file_name())
crime_data = crime_data.withColumn("County",
                                   when(crime_data.file_path.contains("kent"), "Kent")
                                   .when(crime_data.file_path.contains("derbyshire"), "Derbyshire")
                                   .when(crime_data.file_path.contains("leicestershire"), "Leicestershire"))
```

```python
crime_data = crime_data.drop("file_path")
```

```python
crime_data.show(5)
crime_data.printSchema()
```

```python
crime_data.createOrReplaceTempView("crime_data")

total_crimes_by_county = spark.sql("""
    SELECT County, COUNT(*) AS Total_Crimes
    FROM crime_data
    GROUP BY County
""")
total_crimes_by_county.show()
```

```python
monthly_trends = spark.sql("""
    SELECT County, MONTH(Month) AS Month, COUNT(*) AS Total_Crimes
    FROM crime_data
    GROUP BY County, MONTH(Month)
    ORDER BY County, Month
""")
monthly_trends.show()
```

```python
crime_type_ranking = spark.sql("""
    SELECT County, `Crime type`, COUNT(*) AS Crime_Count
    FROM crime_data
    GROUP BY County, `Crime type`
    ORDER BY County, Crime_Count DESC
""")
crime_type_ranking.show(10000)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

monthly_trends_pd = monthly_trends.toPandas()
monthly_trends_pivot = monthly_trends_pd.pivot(index='Month', columns='County', values='Total_Crimes')

monthly_trends_pivot.plot(kind='line', marker='X')
plt.title("Monthly Crime Trends by County")
plt.xlabel("Month")
plt.ylabel("Total Crimes")
plt.legend(title='County')
plt.show()
```

```python
crime_type_ranking_pd = crime_type_ranking.toPandas()
top_crime_types = crime_type_ranking_pd.groupby("County").head(10)  # Adjust number for most common crime types

top_crime_types.pivot(index='Crime type', columns='County', values='Crime_Count').plot(kind='bar', stacked=True)
plt.title("Top Crime Types by County")
plt.xlabel("Crime Type")
plt.ylabel("Frequency")
plt.legend(title='County')
plt.show()
```

```python
for county in crime_type_ranking_pd['County'].unique():
    county_data = crime_type_ranking_pd[crime_type_ranking_pd['County'] == county]
    plt.figure()
    plt.pie(county_data['Crime_Count'], labels=county_data['Crime type'], autopct='%1.1f%%')
    plt.title(f"Crime Type Breakdown for {county}")
    plt.show()
```

```python
from pyspark.sql import SparkSession
from pyspark.ml.feature import StringIndexer, VectorAssembler
from pyspark.sql.functions import col, when, mean, coalesce
from pyspark.sql import functions as F

# Initialize Spark session
spark = SparkSession.builder \
    .appName("Loan Prediction using Spark MLlib") \
    .getOrCreate()

# Load the dataset
train_loan_data = spark.read.csv("/content/LoanData/train_loan.csv", header=True, inferSchema=True)

# Function to handle numeric columns
def fill_numeric_nulls(df, numeric_cols):
    for col_name in numeric_cols:
        median_val = df.select(col_name).na.drop().approxQuantile(col_name, [0.5], 0.01)[0]
        df = df.withColumn(col_name,
                          coalesce(col(col_name), F.lit(median_val)))
    return df

# Define numeric and categorical columns
numeric_cols = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term', 'Credit_History']
categorical_cols = ['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed', 'Property_Area']

# Fill numeric nulls
train_loan_data = fill_numeric_nulls(train_loan_data, numeric_cols)

# Fill categorical nulls with mode values
for cat_col in categorical_cols:
    mode_val = train_loan_data.groupBy(cat_col).count().orderBy('count', ascending=False).first()[0]
    train_loan_data = train_loan_data.fillna({cat_col: mode_val})

# Convert all numeric columns to double type
for num_col in numeric_cols:
    train_loan_data = train_loan_data.withColumn(num_col, col(num_col).cast("double"))

# Create and apply string indexers with handleInvalid="keep"
indexers = [
    StringIndexer(inputCol=col, outputCol=f"{col}_Index", handleInvalid="keep")
    for col in categorical_cols
]

# Apply indexers to the training data
for indexer in indexers:
    model = indexer.fit(train_loan_data)
    train_loan_data = model.transform(train_loan_data)

# Encode the target variable (Loan_Status) in training data
if "Loan_Status" in train_loan_data.columns:
    loan_status_indexer = StringIndexer(inputCol="Loan_Status",
                                      outputCol="Loan_Status_Index",
                                      handleInvalid="keep")
    train_loan_data = loan_status_indexer.fit(train_loan_data).transform(train_loan_data)

# Split the training data into 70/30
train_data, val_data = train_loan_data.randomSplit([0.7, 0.3], seed=42)

# Define feature columns
feature_columns = [f"{col}_Index" for col in categorical_cols] + numeric_cols

# Create VectorAssembler with handleInvalid="keep"
assembler = VectorAssembler(
    inputCols=feature_columns,
    outputCol="features",
    handleInvalid="keep"
)

# Apply VectorAssembler to both train and validation datasets
train_features = assembler.transform(train_data)
val_features = assembler.transform(val_data)

# Select relevant columns
train_final = train_features.select("features", "Loan_Status_Index")
val_final = val_features.select("features", "Loan_Status_Index")  # Keep Loan_Status_Index for evaluation

# Verify the transformations
print("\nTraining Data Sample:")
train_final.show(5, truncate=False)
print("\nValidation Data Sample:")
val_final.show(5, truncate=False)

# Print feature column info
print("\nFeature Columns Used:", feature_columns)


```
